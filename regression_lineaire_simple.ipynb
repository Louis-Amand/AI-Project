# Regression linéaire simple

```python
import pandas as pa
import matplotlib.pyplot as plt
%matplotlib inline
import numpy as np
```

## Exemples (de cours)

### Exemple 1.

```python
data = {'nb_employes': [356, 67, 25, 157, 589, 557, 78],
       'taux_absenteisme': [5, 3, 2, 4, 7, 3, 8]}
```

```python
# On vérifie ses les deux variables sont corrélées (corrcoef de numpy)
```

```python
# Extraction des données
nb_employes = data['nb_employes']
taux_absenteisme = data['taux_absenteisme']

# Calcul du coefficient de corrélation
corr_matrix = np.corrcoef(nb_employes, taux_absenteisme)

# Le coefficient de corrélation entre 'nb_employes' et 'taux_absenteisme'
corr_coefficient = corr_matrix[0, 1]

print("Coefficient de corrélation :", corr_coefficient)
```

### Exemple 2.

#### A la recherche de la meilleure droite de régression 

Nous allons utiliser un corpus généré aléatoirement

```python
from sklearn.datasets import make_regression
X, y = make_regression(n_samples=100, n_features=1, noise=20, random_state=42)
plt.grid()
plt.scatter(X, y)
```

##### Par calcul direct

```python
x_bar = np.mean(X)
y_bar = np.mean(y)
m = len(X)
t1 = [(X[i] - x_bar)*(y[i] - y_bar) for i in range(m)]
t2 = [(X[i] - x_bar)**2 for i in range(m)] 
b1 = np.sum(t1)/np.sum(t2)
b0 = y_bar - b1 * x_bar

print(b0, b1)
y_hat_bis = b0 + b1 * X
```

```python
plt.grid()
plt.scatter(X, y)
# plt.plot(X, y_hat, c='red')
plt.plot(X, y_hat_bis, c='green') ## Ligne droite de regression
```

```python
beta1 = np.sum((X - np.mean(X)) * (y - np.mean(y))) / np.sum((X - np.mean(X))**2)
beta0 = np.mean(y) - beta1 * np.mean(X)
```
